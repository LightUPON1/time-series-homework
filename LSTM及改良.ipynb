```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import seaborn as sns

from statsmodels.graphics.gofplots import qqplot
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.stats.stattools import durbin_watson

from arch import arch_model

plt.rcParams['font.family'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False 

file_path = 'D://2.csv'
data = pd.read_csv(file_path)

data['date'] = pd.to_datetime(data['date'])

# 
data = data.sort_values('date')

# 
data.set_index('date', inplace=True)

# 
plt.figure(figsize=(12, 6))
plt.plot(data['ave'])
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.show()

# 
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['ave'].values.reshape(-1, 1))

def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(look_back, len(dataset)):
        X.append(dataset[i - look_back:i, 0])
        Y.append(dataset[i, 0])
    return np.array(X), np.array(Y)

look_back = 60  

# 
X, Y = create_dataset(scaled_data, look_back)

# 
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

#  LSTM 
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back, 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# 
history = model.fit(X_train, Y_train, epochs=50, batch_size=64, validation_data=(X_test, Y_test), verbose=1)

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='')
plt.plot(history.history['val_loss'], label='')
plt.title('')
plt.ylabel('')
plt.xlabel('')
plt.legend()
plt.show()

# 
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

train_predict = scaler.inverse_transform(train_predict)
Y_train_inv = scaler.inverse_transform(Y_train.reshape(-1, 1))
test_predict = scaler.inverse_transform(test_predict)
Y_test_inv = scaler.inverse_transform(Y_test.reshape(-1, 1))

train_score = np.sqrt(mean_squared_error(Y_train_inv, train_predict))
print(' RMSE: %.2f' % (train_score))
test_score = np.sqrt(mean_squared_error(Y_test_inv, test_predict))
print(' RMSE: %.2f' % (test_score))

train_predict_plot = np.empty_like(scaled_data)
train_predict_plot[:] = np.nan
train_predict_plot[look_back:look_back + len(train_predict), :] = train_predict

test_predict_plot = np.empty_like(scaled_data)
test_predict_plot[:] = np.nan
test_predict_plot[look_back + len(train_predict):look_back + len(train_predict) + len(test_predict), :] = test_predict

# 
plt.figure(figsize=(12, 6))
plt.plot(scaler.inverse_transform(scaled_data), label='', color='blue')
plt.plot(train_predict_plot, label='', color='green')
plt.plot(test_predict_plot, label='', color='red')
plt.title('LSTM ')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()

# 
last_sequence = scaled_data[-look_back:]

# 30
n_future = 30
future_predictions = []

current_sequence = last_sequence.reshape(1, look_back, 1)

for _ in range(n_future):
    future_pred = model.predict(current_sequence)
    future_predictions.append(future_pred[0, 0])
    future_pred_reshaped = future_pred.reshape(1, 1, 1)
    current_sequence = np.concatenate((current_sequence[:, 1:, :], future_pred_reshaped), axis=1)

future_predictions_inv = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

last_date = data.index[-1]
future_dates = [last_date + pd.Timedelta(days=x) for x in range(1, n_future + 1)]

future_df = pd.DataFrame(data=future_predictions_inv, index=future_dates, columns=[''])

print("")
print(future_df)

history_days = 30 
historical_last = data['ave'][-history_days:]

plt.figure(figsize=(12, 6))
plt.plot(historical_last.index, historical_last.values, label=f'{history_days}', color='blue')
plt.plot(future_df.index, future_df[''], label='', marker='o', linestyle='--', color='orange')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()

# 
residuals_train = Y_train_inv.flatten() - train_predict.flatten()
residuals_test = Y_test_inv.flatten() - test_predict.flatten()

residuals = np.concatenate([residuals_train, residuals_test])
residuals_dates = data.index[look_back:look_back + len(residuals)]

residuals_series = pd.Series(residuals, index=residuals_dates)

# 
plt.figure(figsize=(12, 6))
plt.plot(residuals_series.index, residuals_series.values, label='', color='purple')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()

# 
plt.figure(figsize=(12, 6))
sns.histplot(residuals_series, kde=True, bins=30, color='skyblue')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.show()

#  QQ 
plt.figure(figsize=(12, 6))
qqplot(residuals_series, line='s', ax=plt.gca())
plt.title(' QQ ')
plt.show()

#  (ACF) 
plt.figure(figsize=(12, 6))
plot_acf(residuals_series, lags=40, ax=plt.gca())
plt.title(' (ACF) ')
plt.xlabel('')
plt.ylabel('ACF')
plt.show()

# Durbin-Watson 
dw_stat = durbin_watson(residuals_series)
print(f'Durbin-Watson : {dw_stat:.2f}')

# Ljung-Box 
lb_test = acorr_ljungbox(residuals_series, lags=[12], return_df=True)
print("Ljung-Box ")
print(lb_test)


```

```python
#  ARMA(2,2) 
egarch_model = arch_model(
    residuals_series,
    mean='ARX',
    lags=10,  #  lags=3
    vol='EGARCH',
    p=1,
    q=1,
    dist='t'
)

egarch_model_fit = egarch_model.fit(disp='off')
print(egarch_model_fit.summary())

#  eGARCH 
egarch_forecast = egarch_model_fit.forecast(horizon=n_future, method='simulation', simulations=1000)

# 
forecast_variance = egarch_forecast.variance.iloc[-1].mean()

# 
forecast_std = np.sqrt(forecast_variance)

#  LSTM 
final_future_predictions = future_predictions_inv.flatten()

#  future_df
future_df[''] = final_future_predictions

#  95% 
future_df[''] = final_future_predictions + 1.96 * forecast_std
future_df[''] = final_future_predictions - 1.96 * forecast_std

# 
print("")
print(future_df)

# 
plt.figure(figsize=(12, 6))
plt.plot(historical_last.index, historical_last.values, label=f'{history_days}', color='blue')
plt.plot(future_df.index, future_df[''], label='', marker='o', linestyle='--', color='red')
plt.fill_between(future_df.index, future_df[''], future_df[''], color='pink', alpha=0.3, label='95% ')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()

#  adjusted_residuals_df
standardized_residuals = egarch_model_fit.resid / egarch_model_fit.conditional_volatility
standardized_residuals = standardized_residuals.replace([np.inf, -np.inf], np.nan).dropna()

#  DataFrame 
adjusted_residuals_df = pd.DataFrame({'': standardized_residuals}, index=standardized_residuals.index)

# 
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('')

# 
axes[0, 0].plot(adjusted_residuals_df.index, adjusted_residuals_df[''], label='', color='green')
axes[0, 0].set_title('')
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('')
axes[0, 0].legend()
axes[0, 0].grid(True)

# 
sns.histplot(adjusted_residuals_df[''], kde=True, bins=30, color='orange', ax=axes[0, 1])
axes[0, 1].set_title('')
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('')

#  QQ 
qqplot(adjusted_residuals_df[''], line='s', ax=axes[1, 0])
axes[1, 0].set_title(' QQ ')

#  (ACF) 
plot_acf(adjusted_residuals_df[''], lags=40, ax=axes[1, 1])
axes[1, 1].set_title(' (ACF) ')
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel('ACF')
axes[1, 1].grid(True)

# 
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Durbin-Watson 
dw_stat_adj = durbin_watson(adjusted_residuals_df[''])
print(f' Durbin-Watson : {dw_stat_adj:.2f}')

# Ljung-Box 
lb_test_adj = acorr_ljungbox(adjusted_residuals_df[''], lags=[12], return_df=True)
print(" Ljung-Box ")
print(lb_test_adj)
```
