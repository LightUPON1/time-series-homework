```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import seaborn as sns

from statsmodels.graphics.gofplots import qqplot
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.stats.stattools import durbin_watson

from arch import arch_model

plt.rcParams['font.family'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False 

file_path = 'D:/桌面/黄金期货历史数据2.csv'
data = pd.read_csv(file_path)

data['date'] = pd.to_datetime(data['date'])

# 按日期排序
data = data.sort_values('date')

# 设置日期列为索引
data.set_index('date', inplace=True)

# 可视化原始时间序列数据
plt.figure(figsize=(12, 6))
plt.plot(data['ave'])
plt.title('原始时间序列')
plt.xlabel('日期')
plt.ylabel('值')
plt.show()

# 数据预处理：对数据进行缩放
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['ave'].values.reshape(-1, 1))

def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(look_back, len(dataset)):
        X.append(dataset[i - look_back:i, 0])
        Y.append(dataset[i, 0])
    return np.array(X), np.array(Y)

look_back = 60  

# 创建数据集
X, Y = create_dataset(scaled_data, look_back)

# 划分训练和测试集
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# 构建 LSTM 模型
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back, 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# 训练模型
history = model.fit(X_train, Y_train, epochs=50, batch_size=64, validation_data=(X_test, Y_test), verbose=1)

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='训练')
plt.plot(history.history['val_loss'], label='测试')
plt.title('模型损失')
plt.ylabel('损失')
plt.xlabel('迭代次数')
plt.legend()
plt.show()

# 进行预测
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

train_predict = scaler.inverse_transform(train_predict)
Y_train_inv = scaler.inverse_transform(Y_train.reshape(-1, 1))
test_predict = scaler.inverse_transform(test_predict)
Y_test_inv = scaler.inverse_transform(Y_test.reshape(-1, 1))

train_score = np.sqrt(mean_squared_error(Y_train_inv, train_predict))
print('训练集 RMSE: %.2f' % (train_score))
test_score = np.sqrt(mean_squared_error(Y_test_inv, test_predict))
print('测试集 RMSE: %.2f' % (test_score))

train_predict_plot = np.empty_like(scaled_data)
train_predict_plot[:] = np.nan
train_predict_plot[look_back:look_back + len(train_predict), :] = train_predict

test_predict_plot = np.empty_like(scaled_data)
test_predict_plot[:] = np.nan
test_predict_plot[look_back + len(train_predict):look_back + len(train_predict) + len(test_predict), :] = test_predict

# 绘制原始数据和预测值
plt.figure(figsize=(12, 6))
plt.plot(scaler.inverse_transform(scaled_data), label='原始数据', color='blue')
plt.plot(train_predict_plot, label='训练预测', color='green')
plt.plot(test_predict_plot, label='测试预测', color='red')
plt.title('LSTM 模型预测')
plt.xlabel('时间')
plt.ylabel('值')
plt.legend()
plt.grid(True)
plt.show()

# 预测未来值
last_sequence = scaled_data[-look_back:]

# 预测未来30天
n_future = 30
future_predictions = []

current_sequence = last_sequence.reshape(1, look_back, 1)

for _ in range(n_future):
    future_pred = model.predict(current_sequence)
    future_predictions.append(future_pred[0, 0])
    future_pred_reshaped = future_pred.reshape(1, 1, 1)
    current_sequence = np.concatenate((current_sequence[:, 1:, :], future_pred_reshaped), axis=1)

future_predictions_inv = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

last_date = data.index[-1]
future_dates = [last_date + pd.Timedelta(days=x) for x in range(1, n_future + 1)]

future_df = pd.DataFrame(data=future_predictions_inv, index=future_dates, columns=['预测值'])

print("未来预测：")
print(future_df)

history_days = 30 
historical_last = data['ave'][-history_days:]

plt.figure(figsize=(12, 6))
plt.plot(historical_last.index, historical_last.values, label=f'最近{history_days}天的历史数据', color='blue')
plt.plot(future_df.index, future_df['预测值'], label='未来预测', marker='o', linestyle='--', color='orange')
plt.title('近期历史数据与未来预测')
plt.xlabel('日期')
plt.ylabel('值')
plt.legend()
plt.grid(True)
plt.show()

# 计算残差
residuals_train = Y_train_inv.flatten() - train_predict.flatten()
residuals_test = Y_test_inv.flatten() - test_predict.flatten()

residuals = np.concatenate([residuals_train, residuals_test])
residuals_dates = data.index[look_back:look_back + len(residuals)]

residuals_series = pd.Series(residuals, index=residuals_dates)

# 绘制残差随时间的变化
plt.figure(figsize=(12, 6))
plt.plot(residuals_series.index, residuals_series.values, label='残差', color='purple')
plt.title('残差随时间的变化')
plt.xlabel('日期')
plt.ylabel('残差')
plt.legend()
plt.grid(True)
plt.show()

# 绘制残差的直方图和密度图
plt.figure(figsize=(12, 6))
sns.histplot(residuals_series, kde=True, bins=30, color='skyblue')
plt.title('残差的直方图和密度')
plt.xlabel('残差')
plt.ylabel('频率')
plt.show()

# 绘制残差的 QQ 图
plt.figure(figsize=(12, 6))
qqplot(residuals_series, line='s', ax=plt.gca())
plt.title('残差的 QQ 图')
plt.show()

# 绘制残差的自相关函数 (ACF) 图
plt.figure(figsize=(12, 6))
plot_acf(residuals_series, lags=40, ax=plt.gca())
plt.title('残差的自相关函数 (ACF) 图')
plt.xlabel('滞后')
plt.ylabel('ACF')
plt.show()

# Durbin-Watson 检验
dw_stat = durbin_watson(residuals_series)
print(f'Durbin-Watson 统计量: {dw_stat:.2f}')

# Ljung-Box 检验
lb_test = acorr_ljungbox(residuals_series, lags=[12], return_df=True)
print("Ljung-Box 检验：")
print(lb_test)


```

```python
# 使用 ARMA(2,2) 作为均值模型
egarch_model = arch_model(
    residuals_series,
    mean='ARX',
    lags=10,  # 可以调整为更高阶，如 lags=3
    vol='EGARCH',
    p=1,
    q=1,
    dist='t'
)

egarch_model_fit = egarch_model.fit(disp='off')
print(egarch_model_fit.summary())

# 使用 eGARCH 模型预测未来残差
egarch_forecast = egarch_model_fit.forecast(horizon=n_future, method='simulation', simulations=1000)

# 获取预测的条件方差的均值（模拟结果的均值）
forecast_variance = egarch_forecast.variance.iloc[-1].mean()

# 计算预测的标准差
forecast_std = np.sqrt(forecast_variance)

# 使用 LSTM 的预测作为均值
final_future_predictions = future_predictions_inv.flatten()

# 将预测均值添加到 future_df
future_df['调整后的预测值'] = final_future_predictions

# 计算预测区间（例如 95% 置信区间）
future_df['预测上限'] = final_future_predictions + 1.96 * forecast_std
future_df['预测下限'] = final_future_predictions - 1.96 * forecast_std

# 输出调整后的未来预测
print("调整后的未来预测：")
print(future_df)

# 绘制近期历史数据与调整后的未来预测及预测区间
plt.figure(figsize=(12, 6))
plt.plot(historical_last.index, historical_last.values, label=f'最近{history_days}天的历史数据', color='blue')
plt.plot(future_df.index, future_df['调整后的预测值'], label='调整后的未来预测', marker='o', linestyle='--', color='red')
plt.fill_between(future_df.index, future_df['预测下限'], future_df['预测上限'], color='pink', alpha=0.3, label='95% 预测区间')
plt.title('近期历史数据与调整后的未来预测及预测区间')
plt.xlabel('日期')
plt.ylabel('值')
plt.legend()
plt.grid(True)
plt.show()

# 计算并存储标准化残差以生成 adjusted_residuals_df
standardized_residuals = egarch_model_fit.resid / egarch_model_fit.conditional_volatility
standardized_residuals = standardized_residuals.replace([np.inf, -np.inf], np.nan).dropna()

# 创建 DataFrame 存储标准化后的残差
adjusted_residuals_df = pd.DataFrame({'调整后的残差': standardized_residuals}, index=standardized_residuals.index)

# 残差分析的四张图绘制在一张图上
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('调整后残差的分析图')

# 绘制调整后残差随时间的变化
axes[0, 0].plot(adjusted_residuals_df.index, adjusted_residuals_df['调整后的残差'], label='调整后的残差', color='green')
axes[0, 0].set_title('调整后残差随时间的变化')
axes[0, 0].set_xlabel('日期')
axes[0, 0].set_ylabel('残差')
axes[0, 0].legend()
axes[0, 0].grid(True)

# 绘制调整后残差的直方图和密度图
sns.histplot(adjusted_residuals_df['调整后的残差'], kde=True, bins=30, color='orange', ax=axes[0, 1])
axes[0, 1].set_title('调整后残差的直方图和密度')
axes[0, 1].set_xlabel('残差')
axes[0, 1].set_ylabel('频率')

# 绘制调整后残差的 QQ 图
qqplot(adjusted_residuals_df['调整后的残差'], line='s', ax=axes[1, 0])
axes[1, 0].set_title('调整后残差的 QQ 图')

# 绘制调整后残差的自相关函数 (ACF) 图
plot_acf(adjusted_residuals_df['调整后的残差'], lags=40, ax=axes[1, 1])
axes[1, 1].set_title('调整后残差的自相关函数 (ACF) 图')
axes[1, 1].set_xlabel('滞后')
axes[1, 1].set_ylabel('ACF')
axes[1, 1].grid(True)

# 调整子图间距
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Durbin-Watson 检验调整后的残差
dw_stat_adj = durbin_watson(adjusted_residuals_df['调整后的残差'])
print(f'调整后的 Durbin-Watson 统计量: {dw_stat_adj:.2f}')

# Ljung-Box 检验调整后的残差
lb_test_adj = acorr_ljungbox(adjusted_residuals_df['调整后的残差'], lags=[12], return_df=True)
print("调整后残差的 Ljung-Box 检验：")
print(lb_test_adj)
```
